```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd = 'C:\\Users\\THETANG\\anaconda3\\pkgs\\tesseract-4.1.1-had67df9_4\\Library\\bin\\tesseract.exe'

def repath(path = ""):
    curr_path = os.getcwd()
    curr_path_split = curr_path.split("\\")
    new_path = ''
    for i in range(len(curr_path_split)-1):
        new_path = new_path+curr_path_split[i]+'/'
    path_split = path.split("\\")
    for i in range(len(path_split)):
        if i == len(path_split)-1:
            new_path = new_path+path_split[i]
        else:            
            new_path = new_path+path_split[i]+'/'
    return new_path
```

```python
paths = [
    'SRGAN_model/Img_LR.jpg',
    'SRGAN_model/SR.jpeg',
    'Datasets/Train_crop/4-884.jpg',
    'Datasets/Train_crop/6-9650.jpg',
    'Datasets/Train_crop/6848.jpg',
    'Datasets/Train_crop/8849.jpg',
]


custom_config = '-c tessedit_char_whitelist=0123456789 --psm 1 --oem 3'
for path in paths:
    path = repath(path)
    print(path)
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    print("="*30)
    print(pytesseract.image_to_string(img, config=custom_config))
    print("="*30)
# print(pytesseract.image_to_boxes(img))
# cv2.imshow('Result', img)
# cv2.waitKey(0)
```
